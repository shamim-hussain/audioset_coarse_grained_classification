```python
import pandas as pd
import numpy as np

df_annotations = pd.read_csv('annotations.csv.zip').set_index('ytid')
df_annotations
```

```python
df_good = df_annotations.groupby('check_status').get_group('good')
df_good
```

```python
df_good = df_good.drop('check_status', axis=1)
```

```python
df_good.loc[:,['dataset_split','plausible_superclass']].apply(lambda x: tuple(x), axis=1).value_counts()
```

```python
df_good['stratify_criterion'] = df_good.apply(lambda x: x['dataset_split'] + '_' + x['plausible_superclass'], axis=1)
df_good
```

```python
from sklearn.model_selection import train_test_split
train_dataset, test_dataset = train_test_split(df_good, stratify=df_good['stratify_criterion'], test_size=0.35, random_state=777)
```

```python
train_dataset.stratify_criterion.value_counts()
```

```python
test_dataset.stratify_criterion.value_counts()
```

```python
import json

train_ytids = train_dataset.index.values
test_ytids = test_dataset.index.values

with open('train_test_splits/train_ytids.json', 'w') as f:
    json.dump(train_ytids.tolist(), f)

with open('train_test_splits/test_ytids.json', 'w') as f:
    json.dump(test_ytids.tolist(), f)

train_dataset.reset_index().to_csv('train_test_splits/train_dataset.csv.zip',index=False)
test_dataset.reset_index().to_csv('train_test_splits/test_dataset.csv.zip',index=False)
```

```python
from sklearn.model_selection import StratifiedKFold
import pandas as pd

train_annotations = pd.read_csv('../data/train_test_splits/train_dataset.csv.zip').set_index('ytid')
train_annotations
```

```python
split_criterion = train_annotations.apply(lambda x: x['dataset_split'] + '_' + x['plausible_superclass'], axis=1)
split_criterion.value_counts()
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=777)
indices = skf.split(train_annotations, split_criterion.values)

train_annotations['fold'] = -1
for i,(_, test_index) in enumerate(indices):
    train_annotations.iloc[test_index,-1] = i

train_annotations.fold.value_counts()
```

```python
fold_ytids = {}

for i in range(5):
    fold_ytids[str(i)] = list(train_annotations[train_annotations.fold==i].index.values)

for i in range(5):
    print(len(fold_ytids[str(i)]))
```

```python
with open('../data/train_test_splits/train_ytids_folded.json', 'w') as f:
    json.dump(fold_ytids, f)
```
